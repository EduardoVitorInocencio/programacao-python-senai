```python
#--------------------------- Importar as bibliotecas
import xlsxwriter as xlsx
import os

#--------------------------- Caminho e arquivo
caminho = 'C:\\Users\\Aluno\\Documents\\Eduardo-Python\\'
arquivo = 'Exemplo 03.xlsx' #Cria uma nova pasta de trabalho
workbook = xlsx.Workbook(caminho + arquivo)
#---------------------------- Adicionar uma planilha (aba) ao documento

sheetDados = workbook.add_worksheet('Dados')

#----------------------------- Modificando a cor de fundo das células

higherFormat = workbook.add_format({
                                    'bg_color':'#E2EFDA',
                                    'font_color':'green',
                                    'bold': True
                                    })

lowerFormat = workbook.add_format({
                                    'bg_color':'#F8CBAD',
                                    'font_color':'red',
                                    'bold': True
                                    })
#------------------------- Criar um modelo de dados para aplicar a formatação

dados = [
    ['Coluna1','Coluna2','Coluna3','Coluna4'],
    [10,20,30,40],
    [36,87,100,25],
    [78,98,2,5],
    [74,25,63,12]
]


for linha, listas in enumerate(dados):
    sheetDados.write_row(linha,0,listas)
    
#------------------------- Formatação condicional

sheetDados.conditional_format('A1:D5',{'type':'cell',
                                       'criteria':'>=',
                                       'value': 50,
                                       'format':higherFormat
                                      }
                             )

sheetDados.conditional_format('A1:D5',{'type':'cell',
                                       'criteria':'<',
                                       'value': 50,
                                       'format':lowerFormat
                                      }
                             )
    
#------------------------- Fechar o arquivo e salvar
workbook.close()

#------------------------- Abrir o documento

os.startfile(caminho + arquivo)
```
