### Exercício 05 - Formatando e Aplicando Fórmulas

```python
import xlsxwriter as xlsx
import os
#--------------------------------------- Caminho e arquivo
caminho = 'C:\\Users\\Aluno\\Documents\\Eduardo-Python\\'
arquivo = 'Exercício-06.xlsx'
#---------------------------------------- Criar um arquivo e uma aba
workbook = xlsx.Workbook(caminho + arquivo)
sheetDados = workbook.add_worksheet('Dados')
#----------------------------------------- Modificando a cor de fundo
#Formatação Celulas Coluna A
corCabecalhoLinhas = workbook.add_format({
                                'fg_color':'#F4D0C4', # Cor preenchimento
                                'font_color':'#1D3857', # Cor da fonte
                                'bold': True, # Negrito
                                'border': True # Aplicar bordas
                                })

#Formatação Valores
corValores = workbook.add_format({
                                  'num_format': 'R$ #,##0',
                                  'border': True
})

#Formatação dos Totais

corTotais = workbook.add_format({
                                 'num_format': 'R$ #,##0',
                                 'fg_color':'#F4D0C4',
                                 'font_color':'#1D3857',
                                 'bold': True,
                                 'italic': True,
                                 'border': True
})

#Formatação Cabeçalhos Colunas
corCabecalhosColunas = workbook.add_format({
                                             'font_size': 12,
                                             'align': 'center',                                           'bold': True,
                                             'border': True
})

# Formatação Título

corTitulo = workbook.add_format({
                                'bold': True,
                                'font_size': 18,
                                'align': 'center'
})

# Formatação cabeçalhos mesclados
corCabecalhosMesclados = workbook.add_format({
                                'bold': True,
                                'font_size': 12,
                                'align': 'center',
                                'valign': 'center',
                                'border': True
})
#-------------------------------------------------------
sheetDados.merge_range('A1:C1','Controle de Custos', corTitulo)
sheetDados.merge_range('A2:A3','Item', corCabecalhosMesclados)

sheetDados.write('A4','Salário',corCabecalhoLinhas)
sheetDados.write('A5','Horas Extras',corCabecalhoLinhas)
sheetDados.write('A6','Previdência Social',corCabecalhoLinhas)
sheetDados.write('A7','FGTS',corCabecalhoLinhas)
sheetDados.write('A8','Aluguel',corCabecalhoLinhas)
sheetDados.write('A9','Luz/Água/Telefone',corCabecalhoLinhas)
sheetDados.write('A10','Mat. Expediente',corCabecalhoLinhas)
sheetDados.write('A11','Outras Despesas',corCabecalhoLinhas)
sheetDados.write('A12','Total', corCabecalhoLinhas)

sheetDados.merge_range('B2:C2','Total 1º Trimestre', corCabecalhosMesclados)
sheetDados.write('B3','Previsão', corCabecalhosColunas)

sheetDados.write('B4', 2000, corValores)
sheetDados.write('B5', 3150, corValores)
sheetDados.write('B6', 4300, corValores)
sheetDados.write('B7', 5450, corValores)
sheetDados.write('B8', 6600, corValores)
sheetDados.write('B9', 7750, corValores)
sheetDados.write('B10', 8900, corValores)
sheetDados.write('B11', 10050, corValores)

sheetDados.write('C3','Gastos', corCabecalhosColunas)
sheetDados.write('C4', 1550, corValores)
sheetDados.write('C5', 2550, corValores)
sheetDados.write('C6', 3550, corValores)
sheetDados.write('C7', 4550, corValores)
sheetDados.write('C8', 5550, corValores)
sheetDados.write('C9', 6550, corValores)
sheetDados.write('C10', 7550, corValores)
sheetDados.write('C11', 8550, corValores) 

sheetDados.write_formula('B12', '=sum(B4:B11)')
sheetDados.write_formula('C12', '=sum(C4:C11)')
#--------------------------------------------
workbook.close()
#--------------------------------------------
os.startfile(caminho + arquivo)
```
